### Addition Assignment (+ = )

```python
x = 5
x += 3
print(x)
```

### Subtraction Assignment (- =)


```python
x = 5
x -= 3
print(x)
```

### Multiplication Assignment (* =)

```python
x=2
x*=3
print(x)
```

### Division Assignment ( /= )   



```python
x = 6
x /= 3
print(x)
```

### Floor Division Assignment( //= )



```python
x = 6
x//=3
print(x)
```

### Exponentiation Assignment ( **= )

```python
x = 5
x**=3
print(x)
```
